```python
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.DiGraph()
P1 = 'p1'
P2 = 'p2'

G.add_node((1,2))
G.add_node((P1, 'x'))
G.add_edge((1,2), (P1, 'x'))

pos = graphviz_layout(G, prog='dot')
nx.draw(G, pos)
plt.show()
```
